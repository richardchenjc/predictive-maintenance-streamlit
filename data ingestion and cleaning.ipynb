```python
import pandas as pd
import os

# 1. Load the CSV file into a DataFrame
df = pd.read_csv('C:\\Users\\Richard\\Downloads\\Semester break\\Raw\\ai4i2020.csv')

# 2. First look at the data
print("shape of data:", df.shape)
print(df.head())
```

```python
# 3. Rename columns to be Python-friendly
# We map 'Old Name': 'New_Name'
rename_map = {
    'UDI': 'ID',
    'Product ID': 'Product_ID',
    'Type': 'Product_Type',
    'Air temperature [K]': 'Air_Temp',
    'Process temperature [K]': 'Process_Temp',
    'Rotational speed [rpm]': 'Rotational_Speed',
    'Torque [Nm]': 'Torque',
    'Tool wear [min]': 'Tool_Wear',
    'Machine failure': 'Machine_Failure',
    'TWF': 'Tool_Wear_Failure',
    'HDF': 'Heat_Dissipation_Failure',
    'PWF': 'Power_Failure',
    'OSF': 'Overstrain_Failure',
    'RNF': 'Random_Failure'
}

df_clean = df.rename(columns=rename_map)

# Verify the change
print(df_clean.columns)
```

```python
# 4. Check for Nulls and Data Types
df_clean.info()

print("\nMissing Values:")
print(df_clean.isnull().sum())
```

```python
# 5. Save the cleaned data to the processed folder
# index=False means we don't save the row numbers as a column
df_clean.to_csv('C:\\Users\\Richard\\Downloads\\Semester break\\Processed\\ai4i2020_processed.csv', index=False)

print("Data cleaning complete. Saved to processed folder.")
```
